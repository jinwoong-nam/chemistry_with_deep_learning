# Multivariable linear regression

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

torch.manual_seed(1)
```

```python
# define x_train and y_train
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 80],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

# initiate the model
W = torch.zeros((3, 1), requires_grad=True)   # weight
b = torch.zeros(1, requires_grad=True)        # bias

# set the optimizer
optimizer = optim.SGD([W, b], lr=1e-5)        # SGD method, learning rate = 1e-5

nb_epochs = 1000
for epoch in range(nb_epochs + 1):

  # calcluate H(x)
  hypothesis = x_train.matmul(W) + b          # b is broadcasted and summed to each sample

  # calculate cost
  cost = torch.mean((hypothesis - y_train) ** 2)

  # update H(x) using cost
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  print('Epoch {:4d}/{} hypothesis: {} Cost: {:.6f}'.format(
        epoch, nb_epochs, hypothesis.squeeze().detach(), cost.item()))
```
