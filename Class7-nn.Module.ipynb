```python
import torch
import torch.nn as nn
import torch.nn.functional as F

torch.manual_seed(1)
```

# 1. Simple linear regression

```python
# train data
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
# initialize the model using nn module
model = nn.Linear(1,1)   # simple linear regression, input_dim = 1, output_dim = 1
```

```python
# check the parameters of the model
print(list(model.parameters()))
```

```python
# initiate the optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)  # SGD, learning rate=0.01
```

```python
# train the model
nb_epochs = 2000
for epoch in range(nb_epochs+1):

  # H(x)
  prediction = model(x_train)

  # cost
  cost = F.mse_loss(prediction, y_train)  # mse

  optimizer.zero_grad() # initialize gradient to 0
  cost.backward()       # backward calculation
  optimizer.step()      # update weight and bias

  if epoch % 100 == 0:
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()))
```

```python
# test the model with a random input of 4
new_var = torch.FloatTensor([[4.0]])
pred_y = model(new_var)               # forward calculation
print('prediction when input = 4:', pred_y)
```

```python
# check the parameters of the model
print(list(model.parameters()))
```

# 2. Multiple linear regression

```python
torch.manual_seed(1)

# train data
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
# initialize the model using nn module
model = nn.Linear(3,1)   # multiple linear regression, input_dim = 3, output_dim = 1
```

```python
# check the parameters of the model
print(list(model.parameters()))
```

```python
# initiate the optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=1e-5)  # SGD, learning rate=0.01
```

```python
# train the model
nb_epochs = 2000
for epoch in range(nb_epochs+1):

  # H(x)
  prediction = model(x_train)
  # model(x_train) is equivalent to model.forward(x_train)

  # cost
  cost = F.mse_loss(prediction, y_train)  # mse

  optimizer.zero_grad() # initialize gradient to 0
  cost.backward()       # backward calculation
  optimizer.step()      # update weight and bias

  if epoch % 100 == 0:
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()))
```

```python
# test the model with a random input of [73, 80, 75]
new_var = torch.FloatTensor([[73, 80, 75]])
pred_y = model(new_var)               # forward calculation
print('prediction when input = [73, 80, 75]:', pred_y)
```

```python
# check the parameters of the model
print(list(model.parameters()))
```
